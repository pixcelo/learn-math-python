```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# べき乗の計算
def fn(x):
    a = 3
    return x** a # xのa乗
```

```python
x = np.linspace(0,2)
y = fn(x) # y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
