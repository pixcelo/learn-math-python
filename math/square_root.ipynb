

## 平方根

2 乗して x になる数を x の 平方根 (square root) と呼ぶ

正の値と負の値がある (e.g. 9 の平方根は、3 と -3)

$y=\sqrt x$


```python
x = np.linspace(0,9)

# x の正の平方根 x**(1/2)でも同じ
y = np.sqrt(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
