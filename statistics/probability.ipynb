# 確率
ある事象が起きることが期待されている度合い

$P(A)=\frac{a}{n}$

## 余事象
事象Aに対して「Aが起こらないという事象」をAの世事象という

## 確率の収束
- 多くの試行回数を重ねると(事象の発生数/試行数)が確率に収束していく
- サイコロを何度も振って5が出た回数を数えて(5が出た回数/試行数)に変遷を表す

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y = []
total = 0 # 試行数
num_5 = 0 # 5が出た回数
n = 5000 # サイコロを振る回数

for i in range(n):

    # randint(6)は0~5までの乱数を返すので+1する
    if np.random.randint(6)+1 == 5:
        num_5 += 1
    
    total += 1
    x.append(i)
    y.append(num_5/total)
        
```

```python
plt.plot(x, y)
plt.plot(x, [1/6]*n, linestyle='dashed') # y = 1/6がn個入ったリスト

plt.xlabel('x', size=14)
plt.ylabel('y', size=14)
plt.grid()
plt.show()
```
